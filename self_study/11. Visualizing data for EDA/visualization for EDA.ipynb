#  Visualizing the data for EDA

Visualizations are an excellent start to explore data and see relationships between input features

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../data/housing.csv")
df.head()
```

```python
# using pair plot: it plots every column with every column
```

```python
sns.pairplot(df.sample(1000))
```

```python
'''
sns: This is an alias for the Seaborn library, which is a data visualization library built on top of matplotlib.

pairplot(): This is a Seaborn function that creates a matrix of scatter plots and histograms, where each variable in the DataFrame is compared to every other variable.

df: This is a pandas DataFrame that contains the data we want to visualize.

sample(1000): This is a method of the DataFrame that randomly selects 1000 rows from the DataFrame.
'''
```

```python
sns.pairplot(df.sample(1000).drop(["latitude",
                                   "longitude",], axis=1), 
             hue="ocean_proximity")
```

```python
# using kernal density estimate to see how our median values are distributed 
```

```python
for cls in df.ocean_proximity.unique():
    sns.kdeplot(df[df.ocean_proximity==cls].median_house_value, label=cls)
```

```python
sns.jointplot("households", "total_bedrooms" , df)
```

```python
sns.jointplot("households", "total_bedrooms", df)
```

```python
sns.jointplot("population", "total_bedrooms", df, kind="reg")
```

```python
sns.heatmap(df.corr(), square=True)
```

```python
sns.heatmap(df.corr().abs().round(1), square=True, annot=True)
```
