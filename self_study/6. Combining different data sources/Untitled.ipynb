# combining data from different sources

Data comes in different files, so we have to combine them in a single file.

### How to

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/housing.csv")
```

```python
df.head()
```

```python
df_geo = df[["latitude", "longitude", "ocean_proximity"]]
```

```python
# taking geo dataframe
```

```python
df_geo.head()
```

```python
df_geo.join(df["median_house_value"])
```

```python
# taking price dataframe
```

```python
df_price = df[["longitude", "latitude", "median_house_value"]]
```

```python
df_price.head()
```

```python
df_geo.merge(df_price, left_on = "latitude", right_on="latitude")
```

```python
# concatenate two dataframes into one large dataframe
```

```python
pd.concat([df_price, df_geo], join="inner", axis=1)
```
